```python
import pandas as pd
solar_power_dataset=pd.read_csv("solar power by country.csv")
```

```python
solar_power_dataset2=pd.read_csv("solar power by country.csv")
solar_power_dataset3=pd.read_csv("solar power by country.csv")
```

```python
solar_power_dataset.dropna(subset=["2017_New"],axis=0,inplace=True) #inplace modify data frame , axis 0 drop row axis 1 drop column
```

```python
solar_power_dataset
```

```python
solar_power_dataset2
```

```python
import numpy as np
mean=solar_power_dataset2["2016_New"].mean()
solar_power_dataset2["2016_New"].replace(np.nan,mean,inplace=True)
```

```python
solar_power_dataset2
```

```python
solar_power_dataset2["2016_Total"]=solar_power_dataset2["2016_Total"]/1000
```

```python
solar_power_dataset2
```

```python
solar_power_dataset2.rename(columns={"2016_Total":"totals in 2016"},inplace=True)
```

```python
solar_power_dataset2
```

```python
solar_power_dataset2.dtypes
```

```python
solar_power_dataset2["2016_New"]=solar_power_dataset2["2016_New"].astype(int)
```

```python
solar_power_dataset2
```

feature scaling new value=current value/ max value

```python
solar_power_dataset2["2016_New"]=solar_power_dataset2["2016_New"]/solar_power_dataset2["2016_New"].max()
```

```python
solar_power_dataset2
```

min-max method (interpolation)
,new value = current value-min value/range
,range=max-min values

```python
solar_power_dataset2["totals in 2016"]=(solar_power_dataset2["totals in 2016"]-solar_power_dataset2["totals in 2016"].min())/(solar_power_dataset2["totals in 2016"].max()-solar_power_dataset2["totals in 2016"].min())
```

```python
solar_power_dataset2
```

Z score 
new value= current value - mean()/std()

```python
solar_power_dataset2["2019_Total"]=(solar_power_dataset2["2019_Total"]-solar_power_dataset2["2019_Total"].mean())/(solar_power_dataset2["2019_Total"].std())
```

```python
solar_power_dataset2
```

```python
solar_power_dataset5.dropna(subset=["2016_New"],axis=0,inplace=True)
```

```python
solar_power_dataset5
```

```python
import numpy as np
bins=np.linspace(solar_power_dataset5["2016_Total"].min(),solar_power_dataset5["2016_Total"].max(),4)
group_names=["low","medium","high"]
```

```python
solar_power_dataset5["2016_Total_productions"]=pd.cut(solar_power_dataset5["2016_Total"],bins,labels=group_names,include_lowest=True)
```
